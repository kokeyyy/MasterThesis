## 気象データの前処理

```python
data_period = '202001to202309'
```

### 東京

```python
import pandas as pd

raw_df = pd.read_csv('/Users/koki/PycharmProjects/MasterThesis/data/raw/weather_tokyo_' + data_period + '.csv')
raw_df
```

```python
# date_time          = 測定日時
# prec_no            = 地域コード
# block_no           = 地点コード
# pressure_ground    = 気圧(地表)(hPa)
# pressure_sea       = 気圧(海面)(hPa)
# rainfall           = 降水量(mm)
# temperature        = 気温(℃)
# dew_point_humidity = 露点湿度(℃)
# vapor_pressure     = 蒸気圧(hPa)
# humidity           = 湿度(%)
# windspeed_value    = 風速(m/s)
# windspeed_direction = 風向
# sunshine_hours     = 日照時間(h)
# global_solar_radiation = 全天日射量(MJ/m2)
# snowfall          = 降雪(cm)
# snowdepth         = 積雪(cm)
# weather           = 天気
# cloud_amount      = 雲量
# visibility        = 視程(km)
```

```python
cols = ['date_time',
       'rainfall', 'temperature', 'dew_point_humidity',
       'humidity', 'windspeed_value', 'sunshine_hours',
       'global_solar_radiation', 'snowfall', 'weather']
df = raw_df[cols]
df
```

```python
df.isna().sum()
```

```python
# 前後1時間の平均をとってfillna
df['humidity'] = df['humidity'].interpolate()
df['windspeed_value'] = df['windspeed_value'].interpolate()
```

```python
df = df.iloc[range(0, len(df), 3),:]
df[df['weather'].isna()]
```

```python
df.isna().sum()
```

```python
# 1時間後の天気を使ってｆllna
df['weather'] = df['weather'].fillna(method='bfill')
df
```

```python
df.isna().sum()
```

```python
df = pd.get_dummies(df, columns=['weather'])
df = df.reset_index(drop=True)
df = df.add_prefix('東京_')
df
```
