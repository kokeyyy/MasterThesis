## 2つの前処理済みデータを結合して、1つの時系列データに。

```python
import pandas as pd

# import raw dataframe
raw_df1 = pd.read_csv('/Users/koki/PycharmProjects/MasterThesis/data/preprocessed/201601to201912_10areas_for_analysis.csv')
raw_df2 = pd.read_csv('/Users/koki/PycharmProjects/MasterThesis/data/preprocessed/202001to202309_10areas_for_analysis.csv')

print(raw_df1.shape)
print(raw_df2.shape)
```

```python
df = pd.concat([raw_df1, raw_df2])
print(df.shape)
df
```

```python
df.columns
```

```python
df['東京_weather_氷あられ'] = df['東京_weather_氷あられ'].fillna(0).astype(int)
df['東京_weather_氷あられ'].value_counts()
```

```python
df['東京_weather_霧雨'] = df['東京_weather_霧雨'].fillna(0).astype(int)
df['東京_weather_霧雨'].value_counts()
```

```python
df.isna().sum().sum()
```

```python
# use datetime as index
df = df.set_index('datetime', drop=True)
```
