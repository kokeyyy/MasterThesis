```python
data_period = '202001to202309'
```

```python
import pandas as pd

# get n225 data
n225 = pd.read_csv('/Users/koki/PycharmProjects/MasterThesis/data/raw/N225_' + data_period + '.csv')
n225 = n225.iloc[::-1].reset_index(drop=True)
n225
```

```python
date = pd.date_range(start='2020/1/1', end='2023/9/30')  ### Need to set ###
date = date.strftime("%Y-%m-%d")
date = pd.DataFrame(date, columns=['Date'])
n225 = pd.merge(date, n225, on='Date', how='outer')
n225
```

```python
n225['Close'] = n225['Close'].interpolate()
n225
```

```python
n225.isna().sum()
```

```python
n225 = n225.fillna(method='bfill')
n225
```

```python
n225.isna().sum()
```

```python
lst = []
for i in range(n225.shape[0]):
    for _ in range(8):
        lst.append(n225.iloc[i,1])
len(lst)
```

```python
new_n225 = pd.DataFrame(columns=['Date', 'N225_Close'])
date = pd.date_range(start='2020/1/1', end='2023/9/30 21:00', freq='3h')  ### Need to set ###

new_n225.iloc[:,0] = date
new_n225.iloc[:,1] = lst

new_n225
```
